# What are the 5 most popular trips that you would call "commuter trips"?

To answer this question, we first need to decide what a commuter trip is. For simplicity, we'll define a commuter trip as a weekday trip between the morning hours of 6am and 11am, and between the afternoon hours of 3pm and 8pm. This allows us to capture people with work schedules that see them arrive and leave work at abnormal times throughout the day, such as teachers who arrive at work around 7 or 8am and leave work around 3 or 4pm. We can evaluate this in the data. We are also going to classify commuter trips as those that start and end at different stations.

The chart below shows the top 5 trips with how many morning commuter trips were taken from the start station to the end station. The table also shows the ranking of these trips.

```python
# Most popular trips for commuters in morning
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'SELECT \
    distinct(concat(start_station_name, " - " , end_station_name)) as trip_name,\
    count(*) as counts \
    FROM `bigquery-public-data.san_francisco.bikeshare_trips`\
    WHERE EXTRACT(Hour FROM start_date ) >= 6 and EXTRACT(Hour FROM start_date ) < 11 \
    and EXTRACT(DAYOFWEEK FROM start_date) <> 1 and EXTRACT(DAYOFWEEK FROM start_date) <> 7 \
    and start_station_name != end_station_name \
    GROUP BY trip_name ORDER BY counts DESC' > morning_commute.csv
morning_commute = pd.read_csv('morning_commute.csv')
morning_commute['Morn_Rk'] = np.arange(len(morning_commute)) + 1
morning_commute.head()
```

The table below shows the top 5 trips with how many afternoon trips were taken between stations. The table also shows the ranking of these trips.

```python
# Most popular trips for commuters in afternoon
# Reversed trips
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'SELECT \
    distinct(concat(end_station_name, " - " , start_station_name)) as trip_name,\
    count(*) as counts \
    FROM `bigquery-public-data.san_francisco.bikeshare_trips`\
    WHERE EXTRACT(Hour FROM start_date ) >= 15 and EXTRACT(Hour FROM start_date ) < 20 \
    and EXTRACT(DAYOFWEEK FROM start_date) <> 1 and EXTRACT(DAYOFWEEK FROM start_date) <> 7 \
    GROUP BY trip_name ORDER BY counts DESC' > afternoon_commute.csv
afternoon_commute = pd.read_csv('afternoon_commute.csv')
afternoon_commute['Evening_Rk'] = np.arange(len(afternoon_commute)) + 1
afternoon_commute.head()
```

The 5 most popular trips that you would call "commuter trips" can be found by combining the morning and afternoon commute trip counts. This allows us to get the average rank and the full ridership details.

```python
df = morning_commute.merge(afternoon_commute, on='trip_name', how='left')
df['Avg_Rk'] = (df["Morn_Rk"] + df["Evening_Rk"]) * (0.5)
df = df.sort_values(by=["Avg_Rk"], ascending=True)
df['Morning_Count'] = df["counts_x"]
df['Evening_Count'] = df["counts_y"]
df = df.drop(columns=['counts_x',"counts_y"])
df['Ridership'] = df['Morning_Count'] + df['Evening_Count']
df.head(7)
```

By this, it appears that the top 5 most popular commuter trips are.

* 1. Harry Bridges Plaza (Ferry Building) - 2nd at Townsend
* 2. San Francisco Caltrain (Townsend at 4th) - Temporary Transbay Terminal (Howard at Beale)
* 3. San Francisco Caltrain (Townsend at 4th) - Embarcadero at Folsom
* 4. San Francisco Caltrain 2 (330 Townsend) - Townsend at 7th
* 5. Steuart at Market - Embarcadero at Sansome

These have the 5 highest ridership totals among all of the trips between stations. 

These stations are plotted in red below, while the rest of the stations are shown in blue.

```python
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'select *\
from `bigquery-public-data.san_francisco.bikeshare_stations`\
where landmark = "San Francisco"' > mapping.csv

df = pd.read_csv('mapping.csv') 

df1 = df[df['station_id'].isin([51,60,50,55,61,65,69,70,74])]
df2 = df[~df['station_id'].isin([51,60,50,55,61,65,69,70,74])]


BBox = (df.longitude.min(),   df.longitude.max(), df.latitude.min(), df.latitude.max())

import matplotlib.pyplot as plt
ruh_m = plt.imread('SFMap.PNG')
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(df1.longitude, df1.latitude, zorder=1, alpha= 1, c='r', s=20, label="Station in Top 5")
ax.scatter(df2.longitude, df2.latitude, zorder=1, alpha= 1, c='b', s=20, label="Bike Station")
ax.set_title('Bike Locations on San Francisco Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.legend(loc='best')
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
```

In looking at the mapped bikeshare data, it is notable that the locations that are in red, which are those involved in the Top 5 commuter trips. They are near transit hubs like the Ferry Terminal, the CalTrain terminus and the Temporary Transbay Terminal, which makes sense for commuter travel. They also tend to be located around the flatter coastal region.

### More Stations

Next, I wanted to evaluate the percentage of weekday mornings and afternoons where stations run out of bikes. As a former user of this particular bikeshare, I often encounted a problem where stations would either be filled with bikes, meaning that I could not dock a bike there, or a station would not have any bikes, meaning that I could not remove a bike from that location, and would have to walk to the next station (or skip using bikeshare altogether that day). So we will look at the stations on every date and try to see what percentage of days each station had a time where the station had no bikes, meaning that the station was useless for those who wanted to commute from there.

First, we'll look at the mornings at stations.

```python
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'create table `linear-equator-276423.mydataset.morning_bike_dock_status` as\
SELECT station_id, extract(date from time) as day, min(bikes_available) as min_bikes,\
min(docks_available) as min_docks, max(bikes_available) as max_bikes,\
max(docks_available) as max_docks, \
case when min(bikes_available) = 0 then "Zero Bikes" else "Bikes Not Zero" end as bikes_zero,\
 case when min(docks_available) = 0 then "Zero Docks" else "Docks Not Zero" end as docks_zero\
FROM `bigquery-public-data.san_francisco.bikeshare_status`\
WHERE EXTRACT(Hour FROM time ) >= 6 and EXTRACT(Hour FROM time ) < 11 and \
    EXTRACT(DAYOFWEEK FROM time) <> 1 and EXTRACT(DAYOFWEEK FROM time) <> 7 \
group by station_id, day\
order by day'

```

```python
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'select a.station_id, b.name,\
sum(case when bikes_zero = "Zero Bikes" then 1 else 0 end)/count(bikes_zero) as morn_bikes_zero_pct,\
sum(case when docks_zero = "Zero Docks" then 1 else 0 end)/count(docks_zero) as morn_docks_zero_pct\
from  `linear-equator-276423.mydataset.morning_bike_dock_status` a\
left join `bigquery-public-data.san_francisco.bikeshare_stations` b on a.station_id=b.station_id\
group by a.station_id, b.name\
order by morn_bikes_zero_pct desc' > morning_commute_percentages.csv
morn_commute = pd.read_csv('morning_commute_percentages.csv')
morn_commute = morn_commute[morn_commute['morn_bikes_zero_pct'] > 0.2]
morn_commute = morn_commute[morn_commute['station_id'] != 87]

ax = morn_commute.plot(kind='bar',x='name',y='morn_bikes_zero_pct', figsize=[5, 5])

ax.set_xlabel("Name",fontsize=12)
ax.set_ylabel("Percentage",fontsize=12)
ax.set_title("Percentage of Weekday Mornings Where Stations Run Out of Bikes")
plt.show()
```

Next, we'll look at stations in the afternoons and evaluate how often they have no bikes available to rent.

```python
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'create table `linear-equator-276423.mydataset.afternoon_bike_dock_status` as\
SELECT station_id, extract(date from time) as day, min(bikes_available) as min_bikes,\
min(docks_available) as min_docks, max(bikes_available) as max_bikes,\
max(docks_available) as max_docks, \
case when min(bikes_available) = 0 then "Zero Bikes" else "Bikes Not Zero" end as bikes_zero,\
 case when min(docks_available) = 0 then "Zero Docks" else "Docks Not Zero" end as docks_zero\
FROM `bigquery-public-data.san_francisco.bikeshare_status`\
WHERE EXTRACT(Hour FROM time ) >= 15 and EXTRACT(Hour FROM time ) < 20 and \
    EXTRACT(DAYOFWEEK FROM time) <> 1 and EXTRACT(DAYOFWEEK FROM time) <> 7 \
group by station_id, day\
order by day'

```

```python
! bq query --use_legacy_sql=FALSE --format=csv -n 16000 'select a.station_id, b.name,\
sum(case when bikes_zero = "Zero Bikes" then 1 else 0 end)/count(bikes_zero) as eve_bikes_zero_pct,\
sum(case when docks_zero = "Zero Docks" then 1 else 0 end)/count(docks_zero) as eve_docks_zero_pct\
from `linear-equator-276423.mydataset.afternoon_bike_dock_status` a \
left join `bigquery-public-data.san_francisco.bikeshare_stations` b on a.station_id=b.station_id\
group by a.station_id, b.name\
order by eve_bikes_zero_pct desc' > evening_commute_percentages.csv
eve_commute = pd.read_csv('evening_commute_percentages.csv')
eve_commute = eve_commute[eve_commute['eve_bikes_zero_pct'] > 0.2]
eve_commute = eve_commute[eve_commute['station_id'] != 87]

ax = eve_commute.plot(kind='bar',x='name',y='eve_bikes_zero_pct', figsize=[5, 5])
ax.set_xlabel("Name",fontsize=12)
ax.set_ylabel("Percentage",fontsize=12)
ax.set_title("Percentage of Weekday Afternoon Where Stations Run Out of Bikes")
plt.show()
```

Ultimately, I looked at what percentage of days these stations had no bikes available during some portion of the peak commute time. It was a suprisingly large amount of days where users trying to get to work could not count on the bikeshare system. If there are too many days where the bikeshare service remains unreliable, then people will stop using it. About 14 stations had these sort of unavailabilites on 20% or more of weekday mornings, while 16 had these sort of unavailabilites on 20% or more of weekday afternoons. Most of these heavily impacted locations occurred within San Francisco, where the majority of riders use the service. This is something that must be addressed. 

A possible course of action would be to expand these particular locations where there are heavy numbers of outages, in order to try to make it so that users always have a bike that they can rent and a location that they can dock at.

## Offer Discounts at Non-Peak Times

Bikes are only making the company money when they are in use. Specifically, usage by customers (i.e. non-subscribers) is something that generates new money per ride. If there can be pop-up discounts offered at times when these bikes are sitting unused, then the company can potentially increase their profits. We'll evaluate this by looking at both the monthly and daily usage rates of these bicycles. 

First, we'll look at the usage rate of customers and subscribers by month.

```python
#Customer Rides
!bq query --use_legacy_sql=false --format=csv 'SELECT EXTRACT(month FROM start_date) AS month,\
    COUNT(*) as trip_count \
    FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
    WHERE start_station_name != end_station_name and subscriber_type = "Customer" \
    GROUP BY month \
    ORDER BY month' > customer_months.csv 
customer_month = pd.read_csv('customer_months.csv')
customer_month

#Subscriber Rides
!bq query --use_legacy_sql=false --format=csv 'SELECT EXTRACT(month FROM start_date) AS month,\
    COUNT(*) as trip_count \
    FROM `bigquery-public-data.san_francisco.bikeshare_trips` \
    WHERE start_station_name != end_station_name and subscriber_type = "Subscriber" \
    GROUP BY month \
    ORDER BY month' > subscriber_months.csv 
subscriber_month = pd.read_csv('subscriber_months.csv')
subscriber_month

fig = plt.figure(figsize=(18,8))

# Numbers of pairs of bars you want
N = 12

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = customer_month.trip_count
# Specify the values of orange bars (height)
orange_bar = subscriber_month.trip_count

# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, blue_bar , width, label='Customers')
plt.bar(ind + width, orange_bar, width, label='Subscribers')

plt.xlabel("Month")
plt.ylabel("Number of Non-Subscriber Rides")
plt.title("Monthly Non-Subscriber Rides")

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, ['January','February','March','April','May','June','July','August','September','October','November','December'], rotation=45)

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

Notably, there is a distinct decline in ridership from November to April each year. While weather has something to do with this, and while this could potentially be a good time to take some of the bikes out of service for maintenance, it could also be a good time to encourage ridership. The San Francisco Bay Area does not get as cold as many other cities, making biking an option year round. With an average daily high temperature of 60 degrees, according to the [National Weather Service](https://w2.weather.gov/climate/xmacis.php?wfo=mtr), San Francisco is certainly a city where this service could be used more often in the winter months. Thus, it seems very likely that this could be a time to push deals to potential customers.

### Conclusion/Final Reccomendations:

* More Bikes and Docks
 * As mentioned above, there are over a dozen stations that are completely devoid of bikes on 20% of weekday mornings and afternoon during peak commuting times. As the bike share service is dependent on the ease of use, it is imperative that a system have as many bikes and docks in a location as possible. There is no way to know how many rides were missed out on, as a result of the fact that users were unable to use the service when there were no bikes in a location. Additionally, adding extra docks to a station will make it easier for users to return a bike to their intended destination.
* More Stations
 * As previously mentioned, it is important that the bike share system remain easy to use. Thus, if new stations are added in areas with less coverage, then there will be more ridership. This will also make it easier for users to dock their bicycles if a station that they are intending to use has no open docks. Having more stations allows for more users and more ease of use. It will also help ease the burden on some of the most popular stations if placed in nearby hotspots. 
* Extend the number of minutes a bike can be used.
 * With nearly 60,000 trips lasting longer than 20 minutes, it is obvious there is demand for users to be able to take longer trips. There were more than 10,000 trips lasting between 30 and 45 minutes. While there are penalties for users who go above 30 minutes, there are a number of users who still go over the time limit because they want to continue using the service. This means that there should be a system in place to profit off of these types of users. For example, a proposition could be to raise the rental time limit to 1 hour, which would allow users to get more out of their rides and potentially increase use. An alternative idea would be to allow users to pay for a program that allows them to get unlimited longer bike rides, which could allow them to take the bicycle further distances. 
* Programs to encourage off-peak ridership 
 * There are a plethora of times where many of the bikes aren't in use. Thus, as mentioned, a program could be enacted to encourage people to ride bikes in the middle of the day, as a lunchtime special. If the bikeshare program were to partner with local restaurants to offer discounts to users, then they could potentially encourage people to ride a bike to a nearby restaurant to get lunch. Another potential idea for encouraging midday use would be to focus on the group that tends to use the service at this time, namely customers, which could be done by lowering the single-ride prices at these midday times, particularly in the months from November to April, to potentially increase ridership.
